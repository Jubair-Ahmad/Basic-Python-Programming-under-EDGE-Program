```python
import csv
FILE_NAME = "tasks.csv"

# main_menu function
def main_menu():
    print("\nTask Manager with Priority Tracking")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Update Task Status")
    print("4. Sort and Filter Tasks")
    print("5. Exit")
    choice = input("Enter your choice (1-5): ")
    return choice

# add task function
def add_task():
    title = input("Enter Task Title: ")
    description = input("Enter Task Description: ")
    priority = input("Enter Priority (High, Medium, Low): ").capitalize()
    deadline = input("Enter Deadline (format: DD-MM-YYYY): ")
    status = "Pending"  # Default status

    # Save task to CSV
    with open(FILE_NAME, mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([title, description, priority, deadline, status])

    print("Task added successfully!")

# view_tasks function
def view_tasks():
    try:
        with open(FILE_NAME, mode="r") as file:
            reader = csv.reader(file)
            tasks = list(reader)
            if tasks:
                print(f"\n{'Title':<20}{'Description':<30}{'Priority':<10}{'Deadline':<15}{'Status':<10}")
                print("-" * 85)
                for task in tasks:
                    print(f"{task[0]:<20}{task[1]:<30}{task[2]:<10}{task[3]:<15}{task[4]:<10}")
            else:
                print("No tasks found.")
    except FileNotFoundError:
        print("No tasks found. Add a task first.")

# update_task_status function
def update_task_status():
    view_tasks()
    title_to_update = input("Enter the title of the task to update: ")
    new_status = input("Enter new status (Pending, In Progress, Completed): ").capitalize()

    tasks_updated = False
    try:
        with open(FILE_NAME, mode="r") as file:
            reader = csv.reader(file)
            tasks = list(reader)

        with open(FILE_NAME, mode="w", newline="") as file:
            writer = csv.writer(file)
            for task in tasks:
                if task[0] == title_to_update:
                    task[4] = new_status
                    tasks_updated = True
                writer.writerow(task)

        if tasks_updated:
            print("Task status updated successfully!")
        else:
            print("Task not found.")
    except FileNotFoundError:
        print("No tasks found. Add a task first.")

# sort and filter function
def sort_and_filter_tasks():
    try:
        with open(FILE_NAME, mode="r") as file:
            reader = csv.reader(file)
            tasks = list(reader)
            if not tasks:
                print("No tasks to sort or filter.")
                return

            print("\n1. Sort by Priority")
            print("2. Sort by Deadline")
            print("3. Sort by Status")
            print("4. Filter by High Priority")
            print("5. Filter by Pending Tasks")
            choice = input("Enter your choice (1-5): ")

            if choice == "1":
                tasks.sort(key=lambda x: x[2])  # Sort by priority
            elif choice == "2":
                tasks.sort(key=lambda x: x[3])  # Sort by deadline
            elif choice == "3":
                tasks.sort(key=lambda x: x[4])  # Sort by status
            elif choice == "4":
                tasks = [task for task in tasks if task[2] == "High"]
            elif choice == "5":
                tasks = [task for task in tasks if task[4] == "Pending"]
            else:
                print("Invalid choice.")
                return

            print(f"\n{'Title':<20}{'Description':<30}{'Priority':<10}{'Deadline':<15}{'Status':<10}")
            print("-" * 85)
            for task in tasks:
                print(f"{task[0]:<20}{task[1]:<30}{task[2]:<10}{task[3]:<15}{task[4]:<10}")
    except FileNotFoundError:
        print("No tasks found. Add a task first.")

# main function
def main():
    while True:
        choice = main_menu()
        if choice == "1":
            add_task()
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            update_task_status()
        elif choice == "4":
            sort_and_filter_tasks()
        elif choice == "5":
            print("Exiting the Task Manager. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
